```python
import networkx as nx
import random
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
g = nx.Graph()
g.add_node(0)
g.add_node(1)
g.add_edge(0, 1)
```

```python
def agrega_nodo(h):
    pnodes = []
    for n in h.nodes:
        deg = nx.degree(g, n)
        pnodes += [n,] * deg
        
    s = random.choice(pnodes)
    
    t = random.randint(0,999999)
    while t == s:
        t = random.randint(0,999999)
        
    h.add_node(t)
    h.add_edge(s, t)
```

```python
agrega_nodo(g)
nx.draw(g)
```

```python
for n in range(10):
    agrega_nodo(g)
nx.draw(g)
```

```python
for n in range(1000):
    agrega_nodo(g)
nx.draw(g, node_size=1)
```

```python
f = {}
for n in g.nodes:
    deg = nx.degree(g, n)
    if deg in f:
        f[deg] += 1
    else:
        f[deg] = 1
        
plt.plot(f.values())
```

```python
bins, fbins, p = plt.hist([nx.degree(g, n) for n in g.nodes], bins=16)
```
