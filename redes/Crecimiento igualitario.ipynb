```python
import networkx as nx
import random
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
g = nx.Graph()
g.add_node(0)
```

```python
def agrega_nodo(h):
    s = random.choice(list(h.nodes()))
    
    t = random.randint(0,999999)
    while t == s:
        t = random.randint(0,999999)
        
    h.add_node(t)
    h.add_edge(s, t)
```

```python
nx.draw(g)
```

```python
agrega_nodo(g)
nx.draw(g)
```

```python
for n in range(10):
    agrega_nodo(g)
nx.draw(g)
```

```python
for n in range(2000):
    agrega_nodo(g)
nx.draw(g, node_size=1)
```

```python
f = {}
for n in g.nodes:
    deg = nx.degree(g, n)
    if deg in f:
        f[deg] += 1
    else:
        f[deg] = 1
        
plt.plot(f.values())
```

```python
bins, fbins, p = plt.hist([nx.degree(g, n) for n in g.nodes], bins=16)
```
