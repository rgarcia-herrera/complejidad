```python
# importamos bibliotecas para plotear
import matplotlib
import matplotlib.pyplot as plt

# para desplegar los plots en el notebook
%matplotlib inline

import csv
```

```python
noisy = []
with open('seno.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        noisy.append(float(row[2]))
```

```python
alpha = 0.77
smoothed = 0
smooth = []
for v in noisy:
    smoothed = alpha * smoothed + (1 - alpha) * v
    smooth.append(smoothed)
```

```python
plt.plot(noisy, color='lightgrey')
fig = plt.plot(smooth, color='red')
```
